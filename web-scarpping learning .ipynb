```python
with open("smartprix.html","r",encoding="utf-8") as f:
          html=f.read()
```

```python
from bs4 import BeautifulSoup
import numpy as np
```

```python
soup=BeautifulSoup(html,"xml")
```

```python
soup.prettify()
```

```python
# Get all phone containers
containers = soup.find_all("div", {"class": "sm-product has-tag has-features has-actions"})

# Loop through them
for i in containers:
    h2 = i.find("h2")
    if h2:
        print(h2.get_text(strip=True))  # find out the heading name
        print(i.find('span',{'class','price'}).text)  # find out price 

```

```python
# this is the second way  for find out this value 
# Loop through them
for i in containers:
    print(i.find('h2').text) # find out the heading name
    print(i.find('span',{'class':'price'}).text) # find out price 
    print(i.find('div',{'class':'score rank-2-bg'}).find('b').text)  # find out the spec score 
```

```python
# this is the second way  for find out this value 
# Loop through them
for i in containers:
    print(i.find('h2').text) # find out the heading name
    print(i.find('span',{'class':'price'}).text) # find out price 
    try:
        print(i.find('div',{'class':'score rank-2-bg'}).find('b').text)  # find out the spec score 
    except:
        print()   #  it will not through error 
        
```

```python
name=[]
price=[]
specs=[]

for i in containers:
    name.append(i.find('h2').text)
    price.append(i.find('span',{'class':'price'}).text)
    try:
        speces.append(i.find('div',{'class':'score rank-2-bg'}).find('b').text)
    except:
        specs.append(np.nan)
```

```python
containers = soup.find_all('div',{'class':'sm-product has-tag has-features has-actions'})
```

```python
names = []
prices = []
ratings = []
sim = []
processor = []
ram = []
battery = []
display = []
camera = []
card = []
os = []

for i in soup.find_all('div',{'class':'sm-product has-tag has-features has-actions'}):
    try:
        names.append(i.find('h2').text)
    except:
        names.append(np.nan)
    try:
        prices.append(i.find('span',{'class':'price'}).text)
    except:
        price.append(np.nan)
    try:
        ratings.append(i.find('div',{'class':'score rank-2-bg'}).find('b').text)
    except:
        ratings.append(np.nan)
        
    x = i.find('ul',{'class':'sm-feat specs'}).find_all('li')
    try:
        sim.append(x[0].text)
    except:
        sim.append(np.nan)
    try:
        processor.append(x[1].text)
    except:
        processor.append(np.nan)
    try:    
        ram.append(x[2].text)
    except:
        ram.append(np.nan)
    try:
        battery.append(x[3].text)
    except:
        battery.append(np.nan)
    try:
        display.append(x[4].text)
    except:
        display.append(np.nan)
    try:
        camera.append(x[5].text)
    except:
        camera.append(np.nan)
    try:
        card.append(x[6].text)
    except:
        card.append(np.nan)
    try:
        os.append(x[7].text)
    except:
        os.append(np.nan)
```
